```python
Loss_seint = np.load('Loss_seint.npy')
Loss_iseint = np.load('Loss_iseint.npy')
Loss_risgw = np.load('Loss_risgw.npy')
Loss_gw = np.load('Loss_gw.npy')
Loss_egw = np.load('Loss_egw.npy')
```

```python
styles = {
    'SEINT': dict(color='tab:red', linewidth=2, linestyle='-', alpha=1),
    'ISEINT': dict(color='indianred', linewidth=2, linestyle='--', alpha=1),
    'RISGW': dict(color = 'dimgray', linewidth=2, linestyle='--', alpha=1),
    'GW':    dict(color = 'tab:blue', linewidth=2, linestyle='-', alpha=1),
    'EGW':   dict(color = 'deepskyblue', linewidth=2, linestyle='-.', alpha=1),
}


fig, axs = plt.subplots(
    nrows=4, ncols=1, figsize=(8, 5.5),
)

for idx in range(4):
    axs[idx].plot(Loss_risgw[idx],          label='RISGW',   **styles['RISGW'])
    axs[idx].plot(Loss_egw[idx],        label='EGW', **styles['EGW'])
    axs[idx].plot(Loss_gw[idx],        label='GW', **styles['GW'])
    axs[idx].plot(Loss_iseint[idx], label='ISEINT', **styles['ISEINT'])
    axs[idx].plot(Loss_seint[idx],  label='SEINT', **styles['SEINT'])

handles, labels = axs[0].get_legend_handles_labels()
axs[0].legend(
    handles[::-1], labels[::-1],
    loc='upper right',
    ncol=3,
    fontsize=10,
    markerscale=0.8, 
    handlelength=1.5, 
    handletextpad=0.4 
)


axs[0].set_xticks([])
axs[1].set_xticks([])
axs[2].set_xticks([])
axs[3].set_xlabel('Steps')

axs[0].set_title('Original 3D Space', fontsize=12)
axs[1].set_title('First 2D Projection Space', fontsize=12)
axs[2].set_title('Second 2D Projection Space', fontsize=12)
axs[3].set_title('Third 2D Projection Space', fontsize=12)

for ax in axs:
    ax.set_ylim([0,1.2])
    ax.set_yticks([])
    ax.set_xlim([-2, 49])
    ax.set_ylabel('Loss')
    

plt.tight_layout()
plt.show()

```
